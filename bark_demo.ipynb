## Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

preload_models()
```

```python
text_prompt = """
     WOMAN: Star Wars Princess Leia. I'm a member of the imperial senate on a diplomatic mission to Alderon.[emphatic]
     MAN: Darth Vader You are part of the rebel alliance, and a traitor. Take her away.
"""     
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

## Advanced examples

```python
text_prompt = """
    WOMAN: Good night. Sleep tight. Please do not be like Little Bo Peep, who lost her sheep,
         and now she can't fall asleep.
"""
audio_array = generate_audio(text_prompt)
rate = 88000
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    WOMAN: Don't be like Little Bo Peep, who lost her sheep,
         and now she can't fall asleep. [laugh]
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```
